```python
import os
import shutil
import random
import torch

from google.colab import drive
drive.mount('/content/drive')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("[Info] device =", device)

```

```python
random.seed(42)

source_dir = "/content/drive/MyDrive/dataset/PlantDoc/train"
labeled_dir = "/content/drive/MyDrive/dataset/PlantDoc/train_labeled"
unlabeled_dir = "/content/drive/MyDrive/dataset/PlantDoc/train_unlabeled"
```

```python
for class_name in os.listdir(source_dir):
    class_path = os.path.join(source_dir, class_name)

    if not os.path.isdir(class_path):
        continue

    images = os.listdir(class_path)
    random.shuffle(images)

    split_idx = int(0.1 * len(images))  # 10%

    labeled_images = images[:split_idx]
    unlabeled_images = images[split_idx:]

    os.makedirs(os.path.join(labeled_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(unlabeled_dir, class_name), exist_ok=True)

    for img in labeled_images:
        shutil.copy(
            os.path.join(class_path, img),
            os.path.join(labeled_dir, class_name, img)
        )

    for img in unlabeled_images:
        shutil.copy(
            os.path.join(class_path, img),
            os.path.join(unlabeled_dir, class_name, img)
        )

print("Split complete ✅")
```

Check if a class has zero for labeled picture and fixes this issue

```python


for class_name in os.listdir(labeled_path):
    labeled_class_path = os.path.join(labeled_path, class_name)
    unlabeled_class_path = os.path.join(unlabeled_path, class_name)

    labeled_count = len(os.listdir(labeled_class_path))

    if labeled_count == 0:
        unlabeled_images = os.listdir(unlabeled_class_path)

        if len(unlabeled_images) > 0:
            img_to_move = random.choice(unlabeled_images)

            shutil.move(
                os.path.join(unlabeled_class_path, img_to_move),
                os.path.join(labeled_class_path, img_to_move)
            )

            print(f"Moved 1 image to labeled for class: {class_name}")

print("Fix complete ")
```
