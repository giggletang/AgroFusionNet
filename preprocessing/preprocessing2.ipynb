```python
import os
from PIL import Image
from torchvision import transforms
from tqdm import tqdm
```

Weak and Strong Augmentations (baseline 224 by 224 pixels)
Normalizing Pixels


```python
resize_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()   # converts to tensor [0,1]
])

# Weak and Strong for unlabeled images
weak_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor()
])

strong_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(0.4,0.4,0.4,0.1),
    transforms.RandomRotation(15),
    transforms.ToTensor()
])
```

```python
source_dir = "c:/Users/user/OneDrive/Desktop/fall 2025/Spring 2026/PlantDiseaseProject/data/plantvillage"
target_labeled_dir = "c:/Users/user/OneDrive/Desktop/fall 2025/Spring 2026/PlantDiseaseProject/data/plantdoc/train_labeled"
target_unlabeled_dir = "c:/Users/user/OneDrive/Desktop/fall 2025/Spring 2026/PlantDiseaseProject/data/plantdoc/train_unlabeled"
target_test_dir = "c:/Users/user/OneDrive/Desktop/fall 2025/Spring 2026/PlantDiseaseProject/data/plantdoc/test"

output_dir = "c:/Users/user/OneDrive/Desktop/fall 2025/Spring 2026/PlantDiseaseProject/data/"

# Helper function to save tensor as image
def save_tensor_image(tensor, path):
    img = transforms.ToPILImage()(tensor)
    os.makedirs(os.path.dirname(path), exist_ok=True)
    img.save(path)
```

```python
def process_folder(input_dir, output_dir, transform):
    for root, dirs, files in os.walk(input_dir):
        for file in tqdm(files):
            if file.lower().endswith((".png",".jpg",".jpeg")):
                img_path = os.path.join(root, file)
                img = Image.open(img_path).convert("RGB")
                tensor_img = transform(img)  # returns Tensor
                rel_path = os.path.relpath(root, input_dir)
                save_path = os.path.join(output_dir, rel_path, file)
                save_tensor_image(tensor_img, save_path)
```

```python
def process_unlabeled(input_dir, output_weak_dir, output_strong_dir):
    for root, dirs, files in os.walk(input_dir):
        for file in tqdm(files):
            if file.lower().endswith((".png",".jpg",".jpeg")):
                img_path = os.path.join(root, file)
                img = Image.open(img_path).convert("RGB")

                # weak
                weak_img = weak_transform(img)  # tensor
                rel_path = os.path.relpath(root, input_dir)
                save_path_weak = os.path.join(output_weak_dir, rel_path, file)
                save_tensor_image(weak_img, save_path_weak)

                # strong
                strong_img = strong_transform(img)  # tensor
                save_path_strong = os.path.join(output_strong_dir, rel_path, file)
                save_tensor_image(strong_img, save_path_strong)
```

```python
process_folder(source_dir, os.path.join(output_dir,"plantvillage"), resize_transform)

# Labeled Target
process_folder(target_labeled_dir, os.path.join(output_dir,"plantdoc/train_labeled"), resize_transform)

# Test
process_folder(target_test_dir, os.path.join(output_dir,"plantdoc/test"), resize_transform)

# Unlabeled Target
process_unlabeled(
    target_unlabeled_dir,
    os.path.join(output_dir,"plantdoc/train_unlabeled_weak"),
    os.path.join(output_dir,"plantDoc/train_unlabeled_strong")
)
```
